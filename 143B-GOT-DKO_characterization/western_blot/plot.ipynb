```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
import matplotlib.colors as mcolors
palette = list(mcolors.TABLEAU_COLORS.keys())
%matplotlib inline
```

```python
#sns.set_theme(style="ticks", palette="pastel")
sns.set_theme(style="ticks", palette="muted")
sns.set_context("talk")
```

```python
#sns.color_palette(as_cmap=True)
```

```python
df = pd.read_excel('plotting_data.xlsx')
```

```python
df
```

```python
mask = df['Antigen'] == 'ATPase'

g1 = sns.barplot(x="GOT1/2", y="Signal", hue="UCPH", data=df[mask], capsize=.1, errwidth=2, edgecolor='black', linewidth=2.5, alpha=0.7)

g1.set_ylabel('Vinculin norm. signal');
g1.set_xlabel(None);
#g1.legend_.remove()
```

```python

sns_fig = sns.FacetGrid(df, col='Antigen', height=3.5, col_wrap=3, despine=False)
sns_fig.map_dataframe(sns.barplot, x="GOT1/2", y="Signal", hue="UCPH", capsize=.1, errwidth=2, dodge=True, edgecolor='black', linewidth=2, alpha=0.7, palette=palette)
sns_fig.set_axis_labels('GOT1/2', 'Signal (norm. to Vinc.)', size=14)
sns_fig.set_xticklabels(size=14)
sns_fig.set_titles(size=14)
sns_fig.add_legend()
sns_fig.legend.set_title('UCPH (M)')
sns_fig.legend.set_bbox_to_anchor((1.04, 0.5))

sns_fig.tight_layout()
sns_fig.savefig('UCPH_titration_western.pdf', bbox_inches='tight')
```
