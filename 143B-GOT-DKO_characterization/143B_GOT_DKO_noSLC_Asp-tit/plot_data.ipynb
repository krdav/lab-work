```python
import copy
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
import matplotlib.colors as mcolors
palette = list(mcolors.TABLEAU_COLORS.keys())
%matplotlib inline
```

```python
#sns.set_theme(style="ticks", palette="pastel")
sns.set_theme(style="ticks", palette="muted")
sns.set_context("talk")
```

```python
data_file = 'plotting_data.xlsx'
df = pd.read_excel(data_file)
# df = df.sort_values(by=['Rescue', 'Conc'], ascending=[False, True]).reset_index(drop=True)
```

```python
cpal = sns.color_palette()
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(5.4, 4.5))

asp_order = sorted(list(set(df['Asp'])), reverse=True)

g1 = sns.barplot(y='Prlfr', x='Asp', data=df, edgecolor='black', \
                 linewidth=1, capsize=0.08, alpha=0.7, order=asp_order, color=cpal[0])
g1.set_ylabel('Doublings per day');
g1.set_xlabel('Media aspartate (mM)');
handles, labels = g1.get_legend_handles_labels()
labels = ['Vehicle', 'SM']
#g1.legend(loc='upper left', bbox_to_anchor=(1.04, 1), borderaxespad=0, title='Rescue', labels=labels, handles=handles)
g1.grid(True, axis='y')
g1.set_ylim(bottom=-0.5, top=1.15);
g1.set_title('143B GOT DKO')
ax1.axhline(y=0, linestyle='-', zorder=1, linewidth=2.5, color='#444444')
sns.despine()
plt.tight_layout()
fig.savefig('143B_asp-titr.pdf')
```
