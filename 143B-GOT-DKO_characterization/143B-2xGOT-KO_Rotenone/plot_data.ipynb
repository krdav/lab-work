```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
import matplotlib.colors as mcolors
palette = list(mcolors.TABLEAU_COLORS.keys())
%matplotlib inline
```

```python
#sns.set_theme(style="ticks", palette="pastel")
sns.set_theme(style="ticks", palette="muted")
sns.set_context("talk")
```

```python
data_file = 'prlfr_data.xlsx'
df = pd.read_excel(data_file)
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(7.5, 4.5))

mask = (df['Rotenone'] < 110)

g1 = sns.barplot(y='Prlfr', x='Rotenone', hue='Pyr', errorbar=('ci', 95), data=df[mask], capsize=.04, errwidth=2, edgecolor='black', linewidth=2.5, alpha=0.7)

g1.set_ylabel('Doublings per day');
g1.set_xlabel('Rotenone (nM)');
#g1.ticklabel_format(axis='y', style='sci', scilimits=[0, 0])
g1.legend(loc='upper left', bbox_to_anchor=(1.04, 1), borderaxespad=0, title='Pyruvate')
g1.grid(True, axis='y')

plt.tight_layout()
fig.savefig('143B-2xGOT-KO_Rot-tit_Prlfr.pdf')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 4.5))

mask = (df['Rotenone'] < 110)
g1 = sns.lineplot(y='Size', x='Rotenone', hue='Pyr', style='Pyr', errorbar=('ci', 95), data=df[mask], linewidth=2.5, markers=True, dashes=False)

g1.set_ylabel('Cell size (fL)');
g1.set_xlabel('Rotenone (nM)');
#g1.ticklabel_format(axis='y', style='sci', scilimits=[0, 0])
g1.legend(loc='upper left', bbox_to_anchor=(1.04, 1), borderaxespad=0, title='Pyruvate')
g1.grid(True, axis='y')
#g1.set(xscale="log")

plt.tight_layout()
fig.savefig('143B-2xGOT-KO_Rot-tit_size.pdf')
```
