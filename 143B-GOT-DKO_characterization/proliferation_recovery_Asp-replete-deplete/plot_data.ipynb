```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
import matplotlib.colors as mcolors
palette = list(mcolors.TABLEAU_COLORS.keys())
%matplotlib inline
```

```python
#sns.set_theme(style="ticks", palette="pastel")
sns.set_theme(style="ticks", palette="muted")
sns.set_context("talk")
```

```python
data_file = 'results_flat.xlsx'
df = pd.read_excel(data_file)
```

```python
df
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(9, 4))

g1 = sns.barplot(ax=ax1, y='Cells', x='Cell line', hue='Time', ci=95, data=df, capsize=.04, errwidth=2, edgecolor='black', linewidth=2.5, alpha=0.7)

#g1.set_yscale('log')
g1.set_ylabel('Cell count');
#g1.set_xlabel('2xGOT KO clone');
g1.set_xlabel('GOT DKO clone');
g1.ticklabel_format(axis='y', style='sci', scilimits=[0, 0])
g1.legend(loc='upper left', bbox_to_anchor=(1.04, 1), borderaxespad=0, title='Asp depl. (h) ')

plt.tight_layout()
#fig.savefig('Asp-depletion-repletion-recovery.pdf')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4))

mask = df['Cell line'] == '#7'
g1 = sns.barplot(ax=ax1, y='Cells', x='Cell line', hue='Time', ci=95, data=df[mask], capsize=.04, errwidth=2, edgecolor='black', linewidth=2.5, alpha=0.7)

g1.set_ylabel('Cell count');
g1.set_xlabel('GOT DKO clone #7');
g1.ticklabel_format(axis='y', style='sci', scilimits=[0, 0])
g1.get_legend().remove()
g1.set_xticks([])


mask = df['Cell line'] == '#10'
g2 = sns.barplot(ax=ax2, y='Cells', x='Cell line', hue='Time', ci=95, data=df[mask], capsize=.04, errwidth=2, edgecolor='black', linewidth=2.5, alpha=0.7)

g2.set_ylabel('Cell count');
g2.set_xlabel('GOT DKO clone #10');
g2.ticklabel_format(axis='y', style='sci', scilimits=[0, 0])
g2.legend(loc='upper left', bbox_to_anchor=(1.04, 1), borderaxespad=0, title='Asp depl. (h) ')
g2.set_xticks([])




plt.tight_layout()
fig.savefig('Asp-depletion-repletion-recovery.pdf')
```
