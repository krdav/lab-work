```python
for url in url_list:
    #url = "https://hmdb.ca/metabolites/HMDB0000517"
    http = urllib3.PoolManager()
    response = http.request('GET', url)
    soup = BeautifulSoup(response.data)

    # kill all script and style elements
    for script in soup(["script", "style"]):
        script.extract()    # rip it out

    # get text
    text = soup.get_text()

    # break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
    # break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    # drop blank lines
    text = '\n'.join(chunk for chunk in chunks if chunk)
    
    s1 = 'Chemical Formula'
    i1 = text.index(s1)
    s2 = 'Average Molecular Weight'
    i2 = text.index(s2)
    
    print(text[(i1+len(s1)):i2])
    
    
    #break
```

```python
pool_content = pd.read_csv('pool_content.csv', sep='\t')
peaks = pd.read_csv('peaks_P1-9.csv', sep='\t')
```

```python
pool_content
```

```python
pool_content.loc[pool_content['Compound name'] == 'Alanine', 'Pool'].values[0]
```

```python
# Remove U-13C peaks:
sele = peaks['Compound'].str.contains('U-13C')
peaks_std = peaks.loc[~sele, ]
```

```python
polarity = [n[-3:len(n)] for n in peaks_std['Compound']]
cp_name = [n[0:-4] for n in peaks_std['Compound']]
```

```python
peaks_std['Polarity'] = polarity
peaks_std['Compound name'] = cp_name
```

```python
peaks_std
```

```python
in_pool = []
for cp_nam in peaks_std['Compound name'].values:
    P = pool_content.loc[pool_content['Compound name'] == cp_nam, 'Pool'].values[0]
    in_pool.append(P)
```

```python
peaks_std['In_pool'] = in_pool
```

```python
peaks_in_pool = peaks_std[peaks_std['Sample Name'] == peaks_std['In_pool']]
```

```python
peaks_in_pool
```

```python
RT_list = []
Polarity_list = []

for cp_nam in pool_content['Compound name'].values:
    df_tmp = peaks_in_pool.loc[peaks_in_pool['Compound name'] == cp_nam, :]
    max_idx = df_tmp.loc[:, 'Area'].astype('float').idxmax(axis = 1)

    Polarity = df_tmp.loc[max_idx, 'Polarity']
    RT = float(df_tmp.loc[max_idx, 'Actual RT'])
    
    Polarity_list.append(Polarity)
    RT_list.append(RT)
```

```python
pool_content['RT'] = RT_list
pool_content['Polarity'] = Polarity_list
```

```python
pool_content.loc[:, ['Compound name', 'RT', 'Polarity']].to_csv('CP_w_RT_polarity.csv')
```

```python
RT_list = []
Polarity_list = []

for cp_nam in pool_content['Compound name'].values:
    df_tmp = peaks_in_pool.loc[peaks_in_pool['Compound name'] == cp_nam, :]
    
    area_neg = df_tmp.loc[df_tmp.loc[:, 'Polarity'] == 'neg', 'Area'].values[0]
    area_pos = df_tmp.loc[df_tmp.loc[:, 'Polarity'] == 'pos', 'Area'].values[0]
    
    if area_pos > 1.5*area_neg:
        Polarity = 'pos'
        RT = float(df_tmp.loc[df_tmp.loc[:, 'Polarity'] == 'pos', 'Actual RT'].values[0])
    else:
        Polarity = 'neg'
        RT = float(df_tmp.loc[df_tmp.loc[:, 'Polarity'] == 'neg', 'Actual RT'].values[0])

    Polarity_list.append(Polarity)
    RT_list.append(RT)
```

```python
pool_content['RT'] = RT_list
pool_content['Polarity'] = Polarity_list
```
