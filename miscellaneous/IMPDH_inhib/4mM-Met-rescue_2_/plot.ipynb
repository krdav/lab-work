```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
import matplotlib.colors as mcolors
palette = list(mcolors.TABLEAU_COLORS.keys())
%matplotlib inline
```

```python
#sns.set_theme(style="ticks", palette="pastel")
sns.set_theme(style="ticks", palette="muted")
sns.set_context("talk")
```

```python
#sns.color_palette(as_cmap=True)
```

```python
df = pd.read_excel('plotting_data.xlsx')
```

```python
df['Total cell volume (uL)'] = df['Total cell volume (pL)'] /1e6
df.head(3)
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(9, 6))

g1 = sns.barplot(ax=ax1, x="Rescue", y="Count", data=df)

#g1.set_title('')
g1.set_ylabel('Cell count')
g1.set_xlabel('Rescue')
g1.set_yscale("log")
g1.set_xticklabels(g1.get_xticklabels(), rotation=-30, ha='left', minor=False, rotation_mode='anchor');


fig.tight_layout()
fig.savefig('Cell_count.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(9, 6))

g1 = sns.barplot(ax=ax1, x="Rescue", y="Size", data=df)

#g1.set_title('')
g1.set_ylabel('Size (fL)')
g1.set_xlabel('Rescue')
g1.set_xticklabels(g1.get_xticklabels(), rotation=-30, ha='left', minor=False, rotation_mode='anchor');

fig.tight_layout()
fig.savefig('Cell_size.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(9, 6))

g1 = sns.barplot(ax=ax1, x="Rescue", y="Total cell volume (uL)", data=df)

#g1.set_title('')
g1.set_ylabel('Total cell volume (uL)')
g1.set_xlabel('Rescue')
g1.set_xticklabels(g1.get_xticklabels(), rotation=-30, ha='left', minor=False, rotation_mode='anchor');

fig.tight_layout()
fig.savefig('Cell_total-vol.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(9, 6))

g1 = sns.barplot(ax=ax1, x="Rescue", y="Asp-sens signal norm", data=df)

#g1.set_title('')
g1.set_ylabel('Asp-sens (norm. total cell vol)')
g1.set_xlabel('Rescue')
g1.set_yscale("log")
g1.set_xticklabels(g1.get_xticklabels(), rotation=-30, ha='left', minor=False, rotation_mode='anchor');

fig.tight_layout()
fig.savefig('Cell_Asp-sens.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(9, 6))

g1 = sns.barplot(ax=ax1, x="Rescue", y="Asp-sens-RFP", data=df)

#g1.set_title('')
g1.set_ylabel('Asp-sens (norm. Nuc-RFP)')
g1.set_xlabel('Rescue')
g1.set_xticklabels(g1.get_xticklabels(), rotation=-30, ha='left', minor=False, rotation_mode='anchor');

fig.tight_layout()
fig.savefig('Cell_Asp-sens2.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(10, 5))

g1 = sns.scatterplot(ax=ax1, x="Size", y="Count", hue="Rescue", style="Rescue", data=df, s=200)

g1.set_ylabel('Cell count')
g1.set_xlabel('Cell size (fL)')
g1.set_yscale("log")
sns.move_legend(g1, "upper left", bbox_to_anchor=(1, 1))


fig.tight_layout()
fig.savefig('Cell_count-size.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(10, 5))

g1 = sns.scatterplot(ax=ax1, x="Asp-sens signal norm", y="Count", hue="Rescue", style="Rescue", data=df, s=200)

g1.set_ylabel('Cell count')
g1.set_xlabel('Asp-sens (norm. total cell vol)')
g1.set_yscale("log")
#g1.set_xscale("log")
sns.move_legend(g1, "upper left", bbox_to_anchor=(1, 1))


fig.tight_layout()
fig.savefig('Cell_count-Asp.pdf', bbox_inches='tight')
```
