```python
element_natural_prob = {
    'C': 0.0107,
    'H': 0.000115,
    'N': 0.00364
}
```

```python
obs_data = {
    0: 0,
    1: 0,
    2: 0,
    3: 0.0312822,
    4: 2.9254740,
    5: 97.0432438
}
```

```python
element_list = ['C']*5
```

```python
def obj_iso(v1, element_natural_prob, element_list):
    element_prob = [v1[0]]*5

    N_trials = 10000000
    ra = np.random.binomial(1, element_prob, (N_trials, len(element_prob)))
    sim_data = np.bincount(ra.sum(1)) / len(ra) * 100
    loss = abs(obs_data[4] - sim_data[4]) + abs(obs_data[5] - sim_data[5])
    print(v1, loss)
    return(loss)
```

```python
def iso_fun(v1): return(obj_iso(v1, element_natural_prob, element_list))
xopt, fopt = pso(iso_fun, (0.95,), (0.9999999,), swarmsize=20, maxiter=20, minstep=1e-7, minfunc=1e-7)
```
