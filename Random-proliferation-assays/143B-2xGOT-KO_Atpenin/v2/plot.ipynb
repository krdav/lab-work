```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
import matplotlib.colors as mcolors
palette = list(mcolors.TABLEAU_COLORS.keys())
%matplotlib inline
```

```python
#sns.set_theme(style="ticks", palette="pastel")
sns.set_theme(style="ticks", palette="muted")
sns.set_context("talk")
```

```python
df = pd.read_excel('plotting_data.xlsx')
```

```python
df
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

g1 = sns.barplot(ax=ax1, x="Genotype", y="Prlfr", hue="Treatment", data=df, capsize=.1, errwidth=2, edgecolor='black', linewidth=2.5, alpha=0.7)
g1.set_ylabel('Doublings per day');
g1.set_xlabel('143B SLC1A3, GOT');

old_legend = g1.legend_
handles = old_legend.legendHandles
labels = [t.get_text() for t in old_legend.get_texts()]
title = old_legend.get_title().get_text()
g1.legend(handles, labels, title=title, loc='upper left', bbox_to_anchor=(1.04, 1), borderaxespad=0);

fig.tight_layout()
fig.savefig('143B-GOT-DKO_Atpenin-sensitivity.pdf', bbox_inches='tight')
```
