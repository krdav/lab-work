```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
import matplotlib.colors as mcolors
palette = list(mcolors.TABLEAU_COLORS.keys())
%matplotlib inline
```

```python
#sns.set_theme(style="ticks", palette="pastel")
sns.set_theme(style="ticks", palette="muted")
sns.set_context("talk")
```

```python
#sns.color_palette(as_cmap=True)
```

```python
df = pd.read_excel('plotting_data.xlsx')
```

```python
df.head(3)
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(5, 4.4))

g1 = sns.barplot(ax=ax1, x="Rescue", y="Prlfr", data=df, linewidth=1, edgecolor='0.3')

g1.set_title('143B CytB')
g1.set_ylabel('Doublings per day')
g1.set_xlabel('Rescue')
g1.set_xticklabels(g1.get_xticklabels(), rotation=30, ha='right', minor=False, rotation_mode='anchor');
g1.grid(axis='y')
g1.hlines(0, -0.4, 3.4, color='black', alpha=0.6, linewidth=1.7)


fig.tight_layout()
fig.savefig('CytB_D-Lac_Prlfr.pdf', bbox_inches='tight')
```

```python
df_metab = pd.read_excel('plotting_data_metab.xlsx')
df_metab['Cells-Time'] = ['{}-{}'.format(c, t) for c, t in zip(df_metab['Cells'], df_metab['Time'])]
df_metab['Treatment-Time'] = ['{}-{}'.format(c, t) for c, t in zip(df_metab['Treatment'], df_metab['Time'])]
df_metab['Treatment-Cells'] = ['{}-{}'.format(c, t) for c, t in zip(df_metab['Treatment'], df_metab['Cells'])]
```

```python
df_metab.head()
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 4.4))

g1 = sns.barplot(ax=ax1, x="Cells-Time", y="Pyr RR", hue='Treatment', data=df_metab, linewidth=1, edgecolor='0.3')

g1.set_title('143B CytB')
g1.set_ylabel('Pyruvate (RR)')
g1.set_xlabel('Rescue')
#g1.set_xticklabels(g1.get_xticklabels(), rotation=30, ha='right', minor=False, rotation_mode='anchor');
g1.grid(axis='y')
#g1.hlines(0, -0.4, 3.4, color='black', alpha=0.6, linewidth=1.7)
sns.move_legend(g1, "upper left", bbox_to_anchor=(1, 1))


fig.tight_layout()
#fig.savefig('CytB_D-Lac_Prlfr.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 5.5))
x_order = ['20 mM NaCl-0', '20 mM NaCl-48', '20 mM D-Lactate-0', '20 mM D-Lactate-48']


g1 = sns.barplot(ax=ax1, x="Treatment-Time", y="Pyr RR", hue='Cells', \
                 order=x_order, data=df_metab, linewidth=1, edgecolor='0.3')

g1.set_title('143B CytB')
g1.set_ylabel('Pyruvate (RR)')
g1.set_xlabel('Rescue')
g1.set_xticklabels(g1.get_xticklabels(), rotation=30, ha='right', minor=False, rotation_mode='anchor');
g1.grid(axis='y')
#g1.hlines(0, -0.4, 3.4, color='black', alpha=0.6, linewidth=1.7)
sns.move_legend(g1, "upper left", bbox_to_anchor=(1, 1))


fig.tight_layout()
#fig.savefig('CytB_D-Lac_Prlfr.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 5.5))
x_order = ['20 mM NaCl-No', '20 mM NaCl-Yes', '20 mM D-Lactate-No', '20 mM D-Lactate-Yes']


g1 = sns.barplot(ax=ax1, x="Treatment-Cells", y="Pyr RR", hue='Time', \
                 order=x_order, data=df_metab, linewidth=1, edgecolor='0.3')

g1.set_title('143B CytB')
g1.set_ylabel('Pyruvate (RR)')
g1.set_xlabel('Rescue')
g1.set_xticklabels(g1.get_xticklabels(), rotation=30, ha='right', minor=False, rotation_mode='anchor');
g1.grid(axis='y')
#g1.hlines(0, -0.4, 3.4, color='black', alpha=0.6, linewidth=1.7)
sns.move_legend(g1, "upper left", bbox_to_anchor=(1, 1))


fig.tight_layout()
fig.savefig('CytB_D-Lac_Pyr.pdf', bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 5.5))
x_order = ['20 mM NaCl-No', '20 mM NaCl-Yes', '20 mM D-Lactate-No', '20 mM D-Lactate-Yes']


g1 = sns.barplot(ax=ax1, x="Treatment-Cells", y="Lactate IC", hue='Time', \
                 order=x_order, data=df_metab, linewidth=1, edgecolor='0.3')

g1.set_title('143B CytB')
g1.set_ylabel('Lactate (ion count)')
g1.set_xlabel('Rescue')
g1.set_xticklabels(g1.get_xticklabels(), rotation=30, ha='right', minor=False, rotation_mode='anchor');
g1.grid(axis='y')
g1.set_yscale("log")
#g1.hlines(0, -0.4, 3.4, color='black', alpha=0.6, linewidth=1.7)
sns.move_legend(g1, "upper left", bbox_to_anchor=(1, 1))


fig.tight_layout()
#fig.savefig('CytB_D-Lac_Prlfr.pdf', bbox_inches='tight')
```
